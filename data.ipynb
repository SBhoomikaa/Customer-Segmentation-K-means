```python
data2=data[0:150]
```

```python
arr=np.array(data2['CustLocation'])
```

```python
arr
```

```python
from collections import Counter
```

```python
counts=Counter(arr)
```

```python
counts
```

```python
count=[]
arr2=[]
for i in arr:
    if(i not in arr2):
        if(counts[i]>5):
            count.append(counts[i])
            arr2.append(i)
```

```python
import matplotlib.pyplot as plt
```

```python
x=arr2
y=count
```

```python
fig = plt.figure(figsize =(10, 10))
```

```python
plt.pie(count,labels=arr2,autopct='%.2f') 
plt.title("Pie chart")
```

```python
x=arr2
y=count
plt.plot(x,y,"red",marker="x")
plt.grid()
plt.xlabel("Localities")
plt.ylabel("Number of transactions")
plt.title("Variation of no of transactions with localities")
```

```python
plt.bar(arr2, count, color ='darkslateblue',
        width = 0.5)
plt.xlabel("Localities")
plt.ylabel("Number of transactions")
plt.title("Variation of no of transactions with localities")
```
